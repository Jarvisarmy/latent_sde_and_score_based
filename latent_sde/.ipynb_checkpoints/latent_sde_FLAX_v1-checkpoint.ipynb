```python
# output values from 0 to maxval with iters steps
class LinearScheduler(object):
    def __init__(self, iters, maxval=1.0):
        self._iters = max(1,iters)
        self._val = maxval/self._iters
        self._maxval = maxval
    
    def step(self):
        self._val = min(self._maxval, self._val+self._maxval/self._iters)
        
    @property
    def val(self):
        return self._val
```

```python
class EMAMetric(object):
    def __init__(self, gamma: Optional[float]=0.99):
        super(EMAMetric, self).__init__()
        self._val=None
        self._gamma = gamma
    def step(self, x):
        if self._val is None:
            self._val = x
        else:
            self._val = self._gamma * self._val + (1-self._gamma)*x
        return self._val
    
    @property
    def val(self):
        return self._val
```

```python
# make sure that the dvision is stable
def _stable_division(a,b,epsilon=1e-7):
    b = jnp.where(jnp.absolute(b)> epsilon, b, jnp.full_like(b, fill_value=epsilon)*jnp.sign(b))
    return a/b
```

## 5. Simulation of time series

```python
T = 240
t_end = 24.
t_start = 1.
t = np.linspace(t_start,t_end,T)

gap = 160
s = 30
mu = 0
sigma = 0.1
phi = 0.9
def make_time_series():
    z0 = np.random.normal(0,sigma,1)
    z = np.zeros(T)
    z[0] = z0
    shocks = np.random.normal(mu, sigma, T)
    for idx in range(T-1):
        idx_t = t[idx]
        z[idx+1] = phi*z[idx] + shocks[idx] + 0.2*np.sin(2*10*np.pi*(idx_t)/s)
    return z
r = make_time_series()
r_train = r[:gap]
r_test = r[gap:]
plt.plot(range(gap),r_train)
plt.plot(range(gap,T),r_test)
plt.show()
```

```python
def make_data():
    ts = jnp.array(t)
    ts_ext = jnp.array([t_start-0.1] + list(ts) + [t_end+0.1])
    ts_vis = jnp.linspace(t_start,t_end, 1000)
    ys = jnp.array(r[:,None])
    return ts, ts_ext, ts_vis, ys
```

```python


class latentSDE(nn.Module):
    def setup(self, theta=0.5,mu=0.0,sigma=1.0):
        logvar = math.log(sigma**2/(2.*theta))
        self.batch_size = 256
        self.mu = jnp.array([[mu]])
        self.theta = jnp.array([[theta]])
        self.sigma=jnp.array([[sigma]])
        
        # p(y0)
        #self.py0_mean = self.variable('py0_mean',lambda key, shape: jnp.array([[mu]]),(1,1))
        #self.py0_logvar = self.variable('py0_logvar',lambda key, shape: jnp.array([[logvar]]),(1,1))
        self.py0_mean = jnp.array([[mu]])
        self.py0_logvar = jnp.array([[logvar]])
        
        self.layers = [nn.Dense(n) for n in [200,200,1]]
        
        self.qy0_mean = self.param('qy0_mean',lambda key,shape: jnp.array([[mu]]),(1,1))
        self.qy0_logvar = self.param('qy0_logvar', lambda key, shape: jnp.array([[logvar]]), (1,1))
        self.w1 = self.param('w1',lambda key,shape: jnp.zeros(shape),(200,5))
        self.b1 = self.param('b1',lambda key,shape: jnp.zeros(shape),(200,))
        self.w2 = self.param('w2',lambda key,shape: jnp.zeros(shape),(200,200))
        self.b2 = self.param('b2',lambda key,shape: jnp.zeros(shape),(200,))
        self.w3 = self.param('w3',lambda key,shape: jnp.zeros(shape),(1,200))
        self.b3 = self.param('b3',lambda key,shape: jnp.zeros(shape),(1,))
        
    
    def f(self,t, y,args):
        t = jnp.full_like(y, t)
        x = jnp.concatenate((jnp.sin(t),jnp.cos(t),jnp.sin(t**2),jnp.cos(t**2),y),axis=-1)
        def predict(x):
            x = jax.nn.relu(jnp.dot(self.w1,x)+self.b1)
            x = jax.nn.relu(jnp.dot(self.w2,x)+self.b2)
            x = jnp.dot(self.w3,x)+self.b3
            return x
        return vmap(predict)(x)


    def g(self,t, y,args):  # Shared diffusion.
        return jnp.ones((jnp.size(y,0),1))*sigma

    def h(self,t, y,args):  # Prior drift.
        return self.theta * (self.mu - y)

    def f_aug(self,t, y,args):  # Drift for augmented dynamics with logqp term.
        y = y[:, 0:1] # 提取第一列，保持列的形态
        fv, gv, hv = self.f(t, y,args), self.g(t, y,args), self.h(t, y,args)
        u = _stable_division(fv - hv, gv) # 计算u(z,t)
        f_logqp = .5 * (u ** 2).sum(axis=1, keepdims=True) # 计算integral
        return jnp.concatenate([fv, f_logqp], axis=1)

    def g_aug(self,t, y,args):  # Diffusion for augmented dynamics with logqp term.
        y = y[:, 0:1]
        gv = self.g(t, y,args)
        g_logqp = jnp.zeros_like(y)
        return jnp.concatenate([gv, g_logqp], axis=1)
    
    def sample_p(self,ts,eps,control,key):
        print(self.mu)
        t0 = ts[0]
        t1 = ts[-1]
        dt0 = 0.01
        eps_key,bm_key = jrandom.split(key,2)
        y0 = self.py0_mean + eps * self.py0_std
        vf = diffrax.ODETerm(self.h)
        cvf = diffrax.ControlTerm(self.g, control)
        terms = diffrax.MultiTerm(vf,cvf)
        solver = diffrax.Euler()
        saveat=diffrax.SaveAt(ts=ts)
        sol = diffrax.diffeqsolve(terms, solver, t0,t1,dt0,y0=y0,saveat=saveat,adjoint=diffrax.NoAdjoint())
        return sol.ys

    def sample_q(self,ts, params,eps,control,key):
        t0 = ts[0]
        t1 = ts[-1]
        dt0 = 0.01
        eps_key,bm_key = jrandom.split(key,2)
        qy0_mean = params[0]
        qy0_std = get_std(params[1])
        y0 = self.qy0_mean + eps * self.qy0_std
        vf = diffrax.ODETerm(self.f)
        cvf = diffrax.ControlTerm(self.g, control)
        terms = diffrax.MultiTerm(vf,cvf)
        solver = diffrax.Euler()
        saveat=diffrax.SaveAt(ts=ts)
        sol = diffrax.diffeqsolve(terms, solver, t0,t1,dt0,y0=y0,saveat=saveat,adjoint=diffrax.NoAdjoint())
        return sol.ys    
    
    @property
    def py0_std(self):
        return jnp.exp(.5 * self.py0_logvar)

    @property
    def qy0_std(self):
        return jnp.exp(.5 * self.qy0_logvar)
        
    def __call__(self,ts,key):
        init_key,eps_key,bm_key = jrandom.split(key,3)
        
        
        
        t0 = ts[0]
        t1 = ts[-1]
        dt0 = 0.01
        
        eps = jrandom.normal(eps_key,shape=(self.batch_size,1))
        
        y0 = self.qy0_mean + eps*self.qy0_std
    
        logqp0 = jnp.log(self.qy0_std/self.py0_std+(self.py0_std**2+(self.py0_mean-self.qy0_mean)**2)/(2*self.qy0_std**2)-0.5)
        logqp0 = logqp0.sum(axis=1)
        aug_y0 = jnp.concatenate([y0, jnp.zeros((self.batch_size,1))],axis=1)

        control = diffrax.VirtualBrownianTree(t0=t0,t1=t1,tol=dt0/2,shape=(self.batch_size,2),key=bm_key)
        vf = diffrax.ODETerm(self.f_aug)
        cvf = diffrax.ControlTerm(self.g_aug, control)
        terms = diffrax.MultiTerm(vf,cvf)
        solver = diffrax.Euler()
        saveat=diffrax.SaveAt(ts=ts)
        sol = diffrax.diffeqsolve(terms, solver, t0,t1,dt0,y0=aug_y0,saveat=saveat)
        aug_ys = sol.ys
        ys, logqp_path = aug_ys[:, :, 0:1], aug_ys[-1, :, 1]
        logqp = (logqp0+logqp_path).mean()
        
        return ys, logqp
    
```

```python
def update_step(opt, apply_fn,ts, ys,opt_state, params,key):
    def loss_fn(params):
        zs, kl = apply_fn(params,ts,key)
        zs = jnp.squeeze(zs)
        zs = zs[1:-1] 
        logpy = jax.scipy.stats.norm.logpdf(ys,loc=zs,scale=0.0001).sum(axis=0).mean(axis=0)
        loss = -logpy + kl * kl_scheduler.val
        return loss
    loss, grads = jax.value_and_grad(loss_fn)(params)
    update, opt_state = opt.update(grads, opt_state)
    params = optax.apply_updates(params, updates)
    return opt_state, params
```
