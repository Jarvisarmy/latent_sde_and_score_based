### Martingales
* a process $\{S_t,t\in[0,\infty]\}$ is a martingale with respect to the family of information sets $I_t$ and with respect to probability P if for all $t>0$, then
1. $S_t$ is known given $I_t$
2. $E^P[S_t] < \infty$
3. $E^P_t[S_T]=S_t$ for all $t < T$

Mathematically, it means $E_t[S_{t+u}-S_t]=0$. The direction of the future movement in margingales are impossible to forecast

### probability as a measure
* $dP(\overline{z}) = P(\overline{z}-\frac{1}{2}dz_t < z < \overline{z}+\frac{1}{2}dz_t)$
* $\int_{-\infty}^{\infty} dP(z_t) = 1$
* $E(z_t) = \int_{-\infty}^{\infty}z_tdP(z_t)$


### Girsanov's Thm
**some definitions**:
* sample space: all possible outcome
* measure: probabilities of each of the outcomes in the sample space
* previsible: only depend on previous history
* equivalent measure: two measures are equivalent if they have the same space and possibilities

**Thm**:
1. $W_t$ is a Brownian motion with measure P and sample space $\Omega$
2. $X_t$ is a previsible process satisfying $E_p(exp(\frac{1}{2}\int_0^tX_s^2dx)) < \infty$
then exists an equivalent measure Q on $\Omega$ s.t: $\overline{W_t} = W_t + \int_0^tX_sds$ is a brownian motion, this can be written in differential form $d\overline{W_t} = dW_t+X_tdt$
3. $M_t = exp(-\int_0^tX_udW_u-\frac{1}{2}\int_0^tX_u^2du)$
4. by Radon-Nikodym derivative, we know that $dQ = M_tdP$
